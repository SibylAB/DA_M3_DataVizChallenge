```python
import pandas as pd
import matplotlib.pyplot as plt

coffee_df = pd.read_csv("fake_coffee_chain_sales.csv")
coffee_df.head()
```

## Step 2: Review the BAD Charts

Below are two deliberately poor visualizations.  
Your task: Carefully review each and take notes on what you think is wrong, misleading, or just poorly designed.  
Think about chart type, colors, scales, labeling, and whether the chart tells the truth.  


```python
plt.figure(figsize=(10,6))
subset = coffee_df.groupby("Year")["Sales"].sum().reset_index()

# Wrong variable for title, truncated y-axis, misleading color & framing
plt.plot(subset["Year"], subset["Sales"], color="steelblue", linewidth=2, marker="o")
plt.xlabel("Year", fontsize=12, fontweight="bold")
plt.ylabel("Sales", fontsize=12, fontweight="bold")

y_min = subset["Sales"].min()
y_max = subset["Sales"].max()

plt.ylim(
    round(y_min * 0.98),
    round(y_max * 1.02)
)

ax = plt.gca()                      
ax.set_ylim(0.7, ax.get_ylim()[1]) #extreme truncation 

ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.title("Sales Over Time", fontsize=14, pad=20, fontweight="bold", color="black")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6)) 

plt.scatter(
    coffee_df["Profit"], 
    coffee_df["Sales"],
   color = "#2b5ea5",
   alpha = 0.8,
   s= 75) 

plt.xlabel("Sales") 
plt.ylabel("Total Profit") 
plt.title("Overall Coffee Profit Sales") 

ax = plt.gca()
ax.spines["top"].set_visible(False) 
ax.spines["right"].set_visible(False) 

for spine in ax.spines.values(): 
    spine.set_linewidth(0.25)
    spine.set_alpha(0.5)

plt.tight_layout()
plt.show()
```
